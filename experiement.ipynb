```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
### Load the dataset
data = pd.read_csv("Churn_Modelling.csv")
data.head()
```

```python
## drop unwanted columns

data=data.drop(['RowNumber',	'CustomerId',	'Surname'],axis=1)
```

```python
data.head()
```

```python
## encode categorical variables
label=LabelEncoder()
data['Gender']=label.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
## we use One hot encoder
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
encoded_data=encoder.fit_transform(data[['Geography']])
```

```python
encoded_data.toarray()
```

```python
encoder.get_feature_names_out(['Geography'])
```

```python
encoded_data=pd.DataFrame(encoded_data.toarray(),columns=encoder.get_feature_names_out(['Geography']))
```

```python
encoded_data
```

```python
data=pd.concat([data.drop('Geography',axis=True),encoded_data],axis=1)
```

```python
data.head()
```

```python
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label,file)
    
```

```python
with open('oneHot_encoder.pkl','wb') as file:
    pickle.dump(encoder,file)
```

```python
## Divide the independent and dependent features
x= data.drop('Exited',axis=1)
y=data['Exited']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#### ANN implementation 
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense  # Note: Capital 'D' for Dense
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard # Note: 'o' in TensorBoard
```

```python
(x_test.shape[1],)
```

```python
## Buld our model
model=Sequential([
    Dense(64,activation='relu',input_shape=(x_test.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
## compile the model
model.compile(optimizer=opt,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
## Setup the Tensorboard
import datetime
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard

log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callback=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
## set up early stopping
early_stopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
%pip install tensorboard
```

```python
### Train the model
history=model.fit(
    x_train,y_train,validation_data=(x_test,y_test),epochs=100,
    callbacks=[tensorflow_callback,early_stopping]
)
```

```python
model.save('model.h5')
```
